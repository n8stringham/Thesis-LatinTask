```python
filename = 'Augustine_Confessiones.txt'
file2 = 'Augustine-CityofGod-bks1to6.txt'
file3 = 'Augustine-Confessiones2.txt'
```

```python
# Import dependencies 
# Word2Vec
import gensim
from gensim.summarization import textcleaner
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
from gensim.test.utils import datapath
```

```python
import cltk
from cltk.tokenize.latin.sentence import SentenceTokenizer
from cltk.tokenize.word import WordTokenizer
from cltk.stop.latin import STOPS_LIST
```

```python
import re
```

```python
with open(filename) as f_obj:
    contents = f_obj.read().lower()
    # Remove Special Characters and Punctuation
    #cleanedcontents = re.sub("[\*|,|;|:|\[|\]|\.|\(|\)|\}|\{]","", contents)
    
# First tokenize sentences
sent_tokenizer = SentenceTokenizer()
tokenized_sentences = sent_tokenizer.tokenize(contents)

# Next tokenize words within those sentences
words = []
word_tokenizer = WordTokenizer('latin')
# Tokenize words within each sentence
for sentence in tokenized_sentences:
    words.append(word_tokenizer.tokenize(sentence))

words
```

```python
model = Word2Vec(words, size=100, window=10, min_count=5, iter = 20)
```
